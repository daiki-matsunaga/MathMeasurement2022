```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 150

num_i = 1
num_j = 5
num_k = 5
num_l = 2

# 初期にランダムな重み行列を生成
w_ij = np.random.normal(0.0, 1.0, (num_i, num_j))
w_jk = np.random.normal(0.0, 1.0, (num_j, num_k))
w_kl = np.random.normal(0.0, 1.0, (num_k, num_l))

print(f'w_ij =\n {w_ij}\n')
print(f'w_jk =\n {w_jk}\n')
print(f'w_kl =\n {w_kl}\n')

def Sigmoid(u):
    return 1.0/(1.0 + np.exp(-u))

def CalcZi(zl):
    zk = Sigmoid(w_kl@zl)  # layer l -> layer k
    zj = Sigmoid(w_jk@zk)  # layer k -> layer j
    zi = Sigmoid(w_ij@zj)  # layer j -> layer i (output layer)
    return zi
```

```python
# x, yのデータ生成
grid = np.linspace(-1, 1, 51)
lx, ly = np.meshgrid(grid, grid)
lx = lx.flatten()
ly = ly.flatten()
lz = np.array([])

# zlを取り出しziを計算
for x, y in zip(lx, ly):
    zl = np.array([x, y])
    zi = CalcZi(zl)

    lz = np.append(lz, zi)

# グラフに表示
fig, ax = plt.subplots()
ax.contourf(lx.reshape(51, 51), ly.reshape(51, 51), lz.reshape(51, 51), vmin=0, vmax=1)
ax.set_aspect('equal')
plt.show()
```
